```python
from __future__ import print_function
from PIL import Image
import models
import random
import torch
import torchvision.utils as vutils
from torch.autograd import Variable
import torchvision.transforms
```

```python
def convert_img(img_tensor, row_num):
    img_tensor = img_tensor.cpu()
    grid = vutils.make_grid(img_tensor, nrow = row_num, padding=2)
    ndarr = grid.mul(0.5).add(0.5).mul(255).byte().transpose(0, 2).transpose(0, 1).numpy()
    im = Image.fromarray(ndarr)
    return im, ndarr
```

```python
netG = models._netG_1(1, 100, 3, 64, 1) # ngpu, nz, nc, ngf, n_extra_layers
netG.load_state_dict(torch.load( '../results/netG_epoch_99.pth'))
netG.cuda()
netG1 = models._netG_1(1, 100, 3, 64, 1) # ngpu, nz, nc, ngf, n_extra_layers
netG1.load_state_dict(torch.load( '../results/netG_epoch_50.pth'))
netG1.cuda()
```

```python
noise_batch = torch.FloatTensor(36, 100, 1, 1).normal_(0, 1).to('cuda')
noise_batch = Variable(noise_batch)
fake_batch,_ = netG1(noise_batch)
print(fake_batch.shape)
im, _= convert_img(fake_batch.data, 6)
im
```

```python
noise_batch = torch.FloatTensor(36, 100, 1, 1).normal_(0, 1).to('cuda')
noise_batch = Variable(noise_batch)
fake_batch,_ = netG(noise_batch)
print(fake_batch.shape)
im, _= convert_img(fake_batch.data, 6)
im
```

```python
z1 = torch.FloatTensor(6, 100, 1, 1).normal_(0,1)
z2 = torch.FloatTensor(6, 100, 1, 1).normal_(0,1)
dz = (z2 - z1)/6.
z = torch.FloatTensor(36, 100, 1, 1)
for i in range(4):
    z[6*i:6*(i+1),:,:,:] = z1 + i*dz
z = Variable(z)
z_out,_ = netG(z)
z_im, _ = convert_img(z_out.data,6)
z_im
```
